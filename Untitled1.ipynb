```python
import os
print(os.listdir())

import cv2
import numpy as np
import random
import matplotlib.pyplot as plt

from IPython.display import display, Image


fig=plt.figure(figsize=(10, 10))

%matplotlib inline
plt.rcParams['figure.dpi'] = 200
img = cv2.imread('b.jpg',0)
print(img.shape)
plt.subplot(1, 3, 1)
plt.imshow(img)
# display(img)
edges = cv2.Canny(img,100,200)
edges = (255-edges)
ed2 = cv2.adaptiveThreshold(edges,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)

def get_dark_point(im, threshold = 50):
    while True:
        x, y = random.randint(0, im.shape[0]-1), random.randint(0, im.shape[1]-1)
        if im[x,y] < threshold:
            return (x,y)

def get_dp_within(im, p, threshold = 50, dist2 = 2900):
    while True:
        x, y = random.randint(0, im.shape[0]-1), random.randint(0, im.shape[1]-1)
        if (x-p[0])*(x-p[0]) + (y-p[1])*(y-p[1]) < dist2:
            if im[x,y] < threshold:
                return (x,y)


blank_image = np.zeros((img.shape))
blank_image.fill(255)
for i in range(4):
    ps = []
    ps.append(get_dark_point(ed2))
    for j in range(3):
        ps.append(get_dp_within(ed2, ps[0]))
        cv2.line(blank_image, ps[j], ps[j+1], 0, 2)
plt.subplot(1, 3, 2)
plt.imshow(ed2)
# display(ed2)
plt.subplot(1, 3, 3)
plt.imshow(blank_image)
# display(blank_image)
plt.show()

    

    

data:image/png;base64,iVBORw0KGgoAAAANSUhEUgAAAhwAAAFoCAYAAAAcpSI2AAADCUlEQVR4nO3BMQEAAADCoPVPbQo/oAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAICXAd+NAAEwH624AAAAAElFTkSuQmCC
```
